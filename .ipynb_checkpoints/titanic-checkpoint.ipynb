Creatng Transformers

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import OneHotEncoder
```

```python
Now createng anotehr transforemr for habdling categroical values
```

```python
trf1=ColumnTransformer([
('impute_age',SimpleImputer(strategy="median"),[3]),
('imputer_fare',SimpleImputer(strategy="median"),[6]),
('impute_embarked',SimpleImputer(strategy="most_frequent"),[7])
],remainder="passthrough")
```

```python
trf2=ColumnTransformer(
    [
        ('encode_sex',OneHotEncoder(),[2]),
        ('encode_embarked',OneHotEncoder(),[7])
    ],remainder="passthrough"
)
```

```python
X=df.drop("Survived",axis=1)
y=df["Survived"]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.2)
```

NOR CREATING A PIPELINE

MOST IMPORTANT DOUBT WHY DO WE NEED PIPELINE AND HOW IT AVOIDS DATA LEAKAGE
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('model', LogisticRegression())
])

pipeline.fit(X_train, y_train)
# internally it ALWAYS does:
# 1. scale X_train only
# 2. fit model
# CANNOT mess up order! ✅

```python
pipeline=Pipeline(
    [
        ("trf1",trf1),
        ("trf2",trf2),
        ("model",DecisionTreeClassifier())
    ]
)
```

```python
pipeline.fit(X_train,y_train)
prediction=pipeline.predict(X_test)
print(prediction)
```
